```python
import pandas as pd #for dataframe manipulation
import numpy as np #for array manipulation
```

```python
df = pd.read_csv("t20i_info.csv")
df.head()
```

```python
df.isnull().sum() #checking null values
```

```python
df.shape
```

Independent Variables => Batting Team, Bowling Team, City, Current Score, Overs Left, Wickets left, Current run rate, Runs scored in last 5 overs / 
Dependent Variable => Predicted Score

```python
#extracting the city from the venue

df[df.isnull()['city']]['venue'][0].split(" ")[0] #testing
```

```python
# applying the above technique to the full dataset

df['city'] = df['city'].fillna(df['venue'].apply(lambda x: x.split(" ")[0]))
df.head()
```

```python
df.isnull().sum()
```

```python
df['city'].value_counts()
```

```python
# removing cities with low count of matches

eligible_cities = df['city'].value_counts()[df['city'].value_counts()>600].index.to_list() #selecting only cities with 600 or more matches
eligible_cities
```

```python
df = df[df['city'].isin(eligible_cities)]
df.head()
```

```python
df.shape
```

```python
df['current_score'] = df.groupby('match_id')['runs'].cumsum() #getting the cumulative sum of runs
df.head(10)
```

```python
# Calculating number of balls left

df['overs'] = df.ball.apply(lambda x : str(x).split(".")[0]) #creating the overs column
df['balls'] = df.ball.apply(lambda x : str(x).split(".")[1]) #creating the balls column
df.head(20)
```

```python
df['balls_left'] = 120 - (df['overs'].astype(int)*6 + df['balls'].astype(int))
df.head()
```

```python
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x)
df.tail()
```

```python
df['wicket'] = df['player_dismissed'].apply(lambda x:1 if x!='0' else 0)
df.tail()
```

```python
df['wickets_left'] = 10 - df.groupby('match_id')['wicket'].cumsum()
df.tail(120)
```

```python
df['balls_bowled'] = df['overs'].astype(int)*6 + df['balls'].astype(int)
```

```python
df['run_rate'] = (df['current_score']*6)/df['balls_bowled']
df.tail()
```

```python
df.info()
```

```python

#getting the runs scored in the last 5 overs (30 balls)

groups = df.groupby('match_id')

match_id = df['match_id'].unique()
last_five = []
for id in match_id:
    last_five.extend(groups.get_group(id).rolling(window=30)['runs'].sum().values.tolist())
```

```python
df['last_five'] = last_five
df.tail()
```

```python
final_df = df.groupby('match_id')['runs'].sum().reset_index().merge(df, on='match_id')
final_df.head()
```

```python
final_df.columns
```

```python
finalex_df = final_df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','run_rate', 'last_five','runs_x']]
finalex_df
```

```python
finalex_df.dropna(inplace=True)
```

```python
finalex_df
```

```python
finalex_df.isnull().sum()
```

```python
finalex_df.shape
```

```python
# creating dependent variable and independent variables

x = finalex_df.drop(columns =["runs_x"])
y = finalex_df["runs_x"]
```

```python
#creating train and test data

from sklearn.model_selection import train_test_split

x_train,x_test, y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2024)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score,mean_absolute_error
from xgboost import XGBRegressor

```

```python
# label encoding

transformer = ColumnTransformer([
    ('transformer',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
],remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1',transformer),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate = 0.2, max_depth=12,random_state=1))
])
```

```python
pipe.fit(x_train,y_train)
y_pred = pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
import pickle
pickle.dump(pipe,open('pipe.pkl','wb'))
```
